**Importing the necessary libraries**

1) OS library inorder to manipulate the command line to utilize spleeter and ffmpeg

2) Regular Expression library is used to extract the audio file name and discard the path.

```python
import os 
import re
```

**List that holds the names of all processed files**

```python
wav_file_list = [] 
```

**Taking the directory of the unprocessed audio files as input**

And also adding a '/' in the end just in case the user doesn't provide it.

```python
dir_path = input("Enter the directory in which the files exist:")
if not dir_path[-1] == "/": 
    dir_path = dir_path + "/"
```

**Processing the names and the extensions of the files**

The code cell provided below iterates through all the files in the directory and renames them and then appends the file name to the wav_file_list created earlier.

And also coneverts .mp3 or .aac files to .wav .

```python
with os.scandir(dir_path) as entries:
    count = 1 # Keeping the count while naming the files
    for entry in entries:
        name = re.findall(r"'(.*?)'", str(entry), re.DOTALL)[0] # Using re.findall to extract just the name of the file
        new_name1 = dir_path +"test_"+str(count)+name[-4:]
        old_name1 = dir_path+name
        os.rename(old_name1,new_name1)
        if '.mp3' in new_name1 or '.aac' in new_name1:# converting .mp3 and .aac files to .wav
            new_name2 = new_name1[:-3] + "wav"
            print(new_name2+" ******")
            os.system("!ffmpeg -i " + new_name1 +" "+ new_name2)
            wav_file_list.append(new_name2)
            count = count + 1
        elif '.wav' in str(entry) :
            wav_file_list.append(new_name1)
            count = count + 1
```

**Creating a new directory in where the processed files will be stored**

```python
parent_dir = dir_path[:-1] + "_new"
path = os.path.join(parent_dir)
os.mkdir(path)
```

The **split_and_reduce_noise** function utilizes the previously installed tools namely *spleeter* and *ffmpeg* , applies the required effects on the audio files and then outputs the processed audio files.

```python
def split_and_reduce_noise(string):
    dir_path_proto = string[:string.index('/test_')] + "_proto"
    os.system("spleeter separate "+string+ " -o "+ " "+ dir_path_proto)
    dir_path_new = string[:string.index('/test_')] + "_new"
    os.system('ffmpeg -i '+dir_path_proto + string[string.index('/test'):-4]+"/vocals.wav" + ' -af "highpass=f=200, lowpass=f=3000" '+dir_path_new +string[string.index('/test'):])
```

**Using a for loop to call the split_and_reduce_noise function on all the audio files**

```python
for file_ in wav_file_list:
    split_and_reduce_noise(file_)
```
